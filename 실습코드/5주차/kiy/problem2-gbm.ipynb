```python
data = pd.read_csv('/kaggle/input/2021mlfinalexamp2/train.csv')
X_test = pd.read_csv('/kaggle/input/2021mlfinalexamp2/test.csv')
```

```python
X_train = data.iloc[:,:-1]
y_train = data.iloc[:,-1]
```

```python
X_train = X_train.drop(['id','year','month','day'],axis=1)
X_test = X_test.drop(['id','year','month','day'],axis=1)
```

```python
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV
from sklearn.metrics import mean_squared_error, make_scorer
from IPython.display import display
import matplotlib.pyplot as p
```

```python
#Validation function
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(X_train.values)
    rmse= np.sqrt(-cross_val_score(model, X_train.values, y_train, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import xgboost as xgb
import lightgbm as lgb
```

```python
model_gbm = GradientBoostingRegressor(n_estimators=2000,learning_rate=0.05,
                                   max_depth=4, max_features='sqrt', 
                                   loss='huber')
```

```python
model_gbm.fit(X_train,y_train)
```

```python
y_pred=model_gbm.predict(X_train)
```

```python
gbm_rmse = np.sqrt(mean_squared_error(y_train, y_pred))
```
